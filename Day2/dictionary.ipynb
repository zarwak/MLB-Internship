### Calculate average marks of students.

```python
#test case

students_marks = {
    "Alice": 85,
    "Bob": 92,
    "Charlie": 78,
    "Diana": 90,
    "Eve": 88
}

def average_marks(student_marks):
    return sum(student_marks.values())/len(student_marks)

print(average_marks(students_marks))
```

### Count frequency of words in a sentence.

```python
#test case
sentence = "For every good happening, be thankful. For every bad happening, be thankful. For every happening, be thankful."

# Count the frequency of each word
word_count = {}
for word in sentence.split():
    word_count[word] = word_count.get(word, 0) + 1

print(word_count)
```
