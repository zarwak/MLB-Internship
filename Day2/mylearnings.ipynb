# Python Collections Cheat Sheet

Ordered, mutable collections that can hold any data type. They allow duplicates.
## 1. Lists

### Creating Lists

```python


# Empty list
empty = []

# List of integers
numbers = [10, 20, 30, 40]

# Mixed data types
mixed = ["apple", 1.5, True, 42]

# Using the list() constructor
string_list = list("hello")  # Output: ['h', 'e', 'l', 'l', 'o']
```

### Indexing and Slicing
- Indexing gets a single item using list[index].
- Negative indices start from the end, where -1 refers to the last item.
- Slicing gets a sub-list using list[start:stop:step].
- The stop value is exclusive.

```python
fruits = ["apple", "banana", "cherry", "date", "elderberry"]

print(fruits[0])      # apple
print(fruits[-1])     # elderberry
print(fruits[1:4])    # ['banana', 'cherry', 'date'] (stop=4 is exclusive)
print(fruits[::2])    # ['apple', 'cherry', 'elderberry'] (every 2nd)
```

### Adding and Removing Elements

```python
cart = ["milk"]

# Adding
cart.append("eggs")        # Adds to the end: ['milk', 'eggs']
cart.insert(1, "butter")   # Inserts at index 1: ['milk', 'butter', 'eggs']
cart.extend(["bread", "jam"]) # Adds multiple: ['milk', 'butter', 'eggs', 'bread', 'jam']

# Removing
cart.remove("butter")      # Removes the first occurrence
popped = cart.pop()        # Removes AND returns the last item ('jam')
del cart[0]                # Deletes item at index 0 ('milk')
cart.clear()               # Empties the entire list: []
```

### List Methods

```python
nums = [3, 1, 4, 1, 5, 9, 2]

nums.sort()               # Sorts in place: [1, 1, 2, 3, 4, 5, 9]
nums.reverse()            # Reverses in place: [9, 5, 4, 3, 2, 1, 1]
print(nums.index(4))      # Finds index of first 4: 2
print(nums.count(1))      # Counts occurrences of 1: 2

# Copying (IMPORTANT)
new_list = nums.copy()    # Creates a true copy. 
# BEWARE: `new_list = nums` just creates a reference (they point to the same memory)!
```

### List Comprehensions
A concise way to create lists from existing iterables.

Syntax: [expression for item in iterable if condition]

```python
squares = [x**2 for x in range(6)]          # [0, 1, 4, 9, 16, 25]
evens = [x for x in range(10) if x % 2 == 0] # [0, 2, 4, 6, 8]
nested_loop = [x+y for x in [1,2] for y in [10,20]] # [11, 21, 12, 22]
```

## 2. Tuples
Ordered, immutable collections. They are like lists but write-protected.

### Creating Tuples

```python
empty = ()
single = (5,)          # Note the comma! (5) is just an integer.
colors = ("red", "green", "blue")
mixed = (1, "hello", 3.14)

# Using tuple() constructor
from_list = tuple([1, 2, 3])  # (1, 2, 3)
```

### Tuple Operations
You can index, slice, and concatenate tuples, but you cannot modify them.

```python
a = (1, 2, 3)
b = (4, 5)
print(a[0])        # 1
print(a + b)       # (1, 2, 3, 4, 5) - creates a NEW tuple
print(b * 3)       # (4, 5, 4, 5, 4, 5)
print(len(a))      # 3
print(2 in a)      # True
```

## 3. Sets
Unordered, mutable collections of unique items. They are optimized for fast membership checking.

### Creating Sets

```python
empty = set()          # NOT {}, that's a dictionary!
numbers = {1, 2, 3, 4}
mixed = {1, "hello", 3.14}  # Duplicates are automatically ignored.
from_list = set([1, 2, 2, 3]) # {1, 2, 3}
```

### Removing Duplicates
This is the classic use case for sets.

```python
names = ["Alice", "Bob", "Alice", "Charlie", "Bob"]
unique_names = set(names)    # {'Alice', 'Bob', 'Charlie'}
# If you need it back as a list: list(unique_names)
```

### Set Operations
You can add or remove items from a set.

```python
s = {1, 2, 3}
s.add(4)          # {1, 2, 3, 4}
s.add(2)          # No change (2 already exists)
s.remove(3)       # Removes 3. Raises KeyError if not found.
s.discard(5)      # Removes 5 if exists, DOES NOT raise error (safer).
popped = s.pop()  # Removes and returns an arbitrary element.
```

### Union, Intersection, and Difference
These operations make sets especially powerful for comparisons.

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

# Union - All items in either set
print(A | B)        # {1, 2, 3, 4, 5, 6}
print(A.union(B))   # same

# Intersection - Items in BOTH sets
print(A & B)        # {3, 4}
print(A.intersection(B))

# Difference - Items in A but NOT in B
print(A - B)        # {1, 2}
print(A.difference(B))

# Symmetric Difference - Items in A or B, but NOT both
print(A ^ B)        # {1, 2, 5, 6}
print(A.symmetric_difference(B))
```

## 4. Dictionaries
Unordered but insertion-ordered in Python 3.7+, mutable collections of key-value pairs. Keys must be unique and immutable, while values can be anything.

### Key-Value Pairs

```python
# Creating a dictionary
student = {
    "name": "Alice",
    "age": 22,
    "grades": [85, 90, 78]
}

# Using dict() constructor
car = dict(brand="Toyota", model="Corolla", year=2020)
```

### Accessing and Updating Values

```python
# Access (safe way vs risky way)
print(student["name"])        # Alice (Raises KeyError if key missing)
print(student.get("age"))     # 22 (Returns None if key missing, safe!)
print(student.get("city", "Unknown")) # Returns "Unknown" if missing.

# Updating
student["age"] = 23           # Update existing key
student["city"] = "New York"  # Add new key-value pair
```

### Dictionary Methods

```python
d = {"a": 1, "b": 2, "c": 3}

# Get all keys, values, or items (key-value pairs)
print(d.keys())   # dict_keys(['a', 'b', 'c'])
print(d.values()) # dict_values([1, 2, 3])
print(d.items())  # dict_items([('a', 1), ('b', 2), ('c', 3)])

# Merging two dicts (Python 3.9+)
d2 = {"d": 4, "e": 5}
d.update(d2)      # d now has {'a':1, 'b':2, 'c':3, 'd':4, 'e':5}

# Removing items
removed_value = d.pop("c")   # Removes 'c' and returns 3.
last_item = d.popitem()      # Removes and returns the last inserted item (('e', 5)).
```

### Nested Dictionaries
These are dictionaries inside other dictionaries, useful for representing structured data.

```python
company = {
    "name": "TechCorp",
    "departments": {
        "HR": {
            "manager": "Bob",
            "employees": 5
        },
        "IT": {
            "manager": "Alice",
            "employees": 10,
            "projects": ["Alpha", "Beta"]
        }
    }
}

# Accessing nested data
print(company["departments"]["IT"]["manager"])    # Alice
print(company["departments"]["IT"]["projects"][0]) # Alpha

# Updating nested data
company["departments"]["HR"]["employees"] = 6
company["departments"]["IT"]["projects"].append("Gamma") # Adds to the list inside
```
