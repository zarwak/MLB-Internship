### Count occurrences of an element.

```python
# test cases
tpl = (1,3,5,7,9,3,1)

# built-in function
print(tpl.count(1))

#custom function without using built-in function
def count_occurrences(tpl, value):
    count = 0
    for item in tpl:
        if item == value:
            count += 1
    return count

# test the custom function
print(count_occurrences(tpl, 1))
```

### Convert a tuple into a list and vice versa.

```python
tpl = (1,3,5,7,9,3,1)
lst = [1,3,5,7,9,3,1]

to_list = tuple(lst)
to_tuple = list(tpl)

print(to_list)  
print(to_tuple)
```
