### Find the largest number in a list.

```python
#test case
lst=[1,5,3,2,2,1,4]

#custom function with my logic
def maximum(lst):
    if not lst:
        return None
    max = lst[0]
    for i in range(1,len(lst)):
        if lst[i] > max:
            max = lst[i]
    return max

#function-calling
print(maximum(lst))

#built-in function
print(max(lst))
```

### Find the second largest number.

```python
#test case
lst=[1,5,3,2,2,1,4]

def second_max(lst):
    if len(lst) <2:
        return None
    first = second = float('-inf')
    for n in lst:
        if n > first:
            second = first
            first = n
        elif n > second and n != first:
            second = n
    return second

#function-calling
print(second_max(lst))
```

### Remove duplicate values from a list.

```python
#test case
lst=[1,5,3,2,2,1,4]

#built-in functions/methods
print(list(set(lst)))  # Remove duplicates

#or

#custom function with my logic
def remove_duplicates(lst):
    my_set = set()
    result = []
    for item in lst:
        if item not in my_set:
            my_set.add(item)
            result.append(item)
    return result

#function-calling
print(remove_duplicates(lst))
```

### Reverse a list without using built-in reverse().

```python
#test case
lst = [5,3,4,8,2,1]

#slicing method
lst1= lst[::-1]

# custom function with my logic
def reverse_list(lst):
    left = 0
    right = len(lst) - 1
    while left < right:
        lst[left], lst[right] = lst[right], lst[left]
        left += 1
        right -= 1
    return lst

print(lst1)
print(reverse_list(lst))
```

### Find common elements between two lists.

```python
#test cases
lst1 = [1, 2, 9, 4, 5]
lst2 =  [5,3,4,8,2,1]

# built-in functions/methods
print(list(set(lst1) & set(lst2)))  # Intersection

def common_elements(lst1, lst2):
    lst=[]
    for item in lst1:
        if item in lst2 and item not in lst:
            lst.append(item)
    return lst

print(common_elements(lst1, lst2))
```
