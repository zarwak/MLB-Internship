### Find unique values from a list.

### Perform union and intersection operations.

```python
#test case
s1 = {1, 3, 5, 7, 9, 3, 1}
s2 = {2, 4, 6, 8, 10}

print(s1.union(s2))  
print(s1.intersection(s2)) 

print(s1 | s2)
print(s1 & s2)
```
