## Conditional Statements



Check whether a number is positive, negative, or zero.


```python
#test cases
nums = [4,-1,0]

#function
def type(num):
    if num > 0: return "Positive"
    elif num < 0: return "Negative"
    else: return "Zero"

def main():
    for num in nums:
        print(type(num))
if __name__ == "__main__":
    main()
```

Check whether a number is even or odd.


```python
#test cases
nums = [4,3,5,0]

def is_even(num):
    if num % 2 == 0: return True
    else: return False

def main():
    for num in nums:
        print(is_even(num))
if __name__ == "__main__":
    main()
```

Create a grade calculator based on marks.


```python
#test cases
nums = [74,81,60,45]

def grade(num):
    if num >= 90: return "A"
    elif num >= 80: return "B"
    elif num >= 70: return "C"
    elif num >= 60: return "D"
    else: return "F"

def main():
    for num in nums:
        print(grade(num))

if __name__ == "__main__":
    main()
```

Check whether a year is a leap year.



```python
#test cases
years=[2029,2075,1999,2000,2001,2004]

def leap_year(year):
    if year % 4 == 0 and year % 100 != 0 or year % 400 == 0:
        return True
    else:
        return False

def  main():
    for year in years:
        print(leap_year(year))
main()
```

Count the number of digits in a number.



```python
num = int(input("Enter a number: "))
def digit_count(num):
    count = 0
    while num > 0:
        num //= 10
        count += 1
    return count

def digit_count2(num):
    return len(str(num))

print(digit_count(num))
print(digit_count2(num))
```

Logic Building Problems



Reverse a number.


```python
def reverse_num(num):
    rev = 0
    while num > 0:
        rev = rev * 10 + num % 10
        num //= 10
    return rev

num = int(input("Enter a number: "))
print(reverse_num(num))
```

Check whether a number is a palindrome.


```python
def is_palindrome(num):
    def reverse_num(num):
        rev = 0
        while num > 0:
            rev = rev * 10 + num % 10
            num //= 10
        return rev
    return num == reverse_num(num)

def main():
    num = int(input("Enter a number: "))
    if is_palindrome(num):
        print(f"{num} is a palindrome")
    else:
        print(f"{num} is not a palindrome")
main()
```

Generate the Fibonacci sequence.

```python
def fibonacci(n):
    if n <= 0:
        return []
    elif n == 1:
        return [0]
    elif n == 2:
        return [0, 1]
    else:
        fib = [0, 1]
        for i in range(2, n):
            fib.append(fib[i-1] + fib[i-2])
        return fib
def main():
    n = int(input("Enter the number of terms: "))
    print(fibonacci(n)) 
main()

```

Check whether a number is prime.

```python
#test cases
nums = [4,3,5,0]

def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def main():
    for num in nums:
        print(is_prime(num))
main()
```

Find all prime numbers between 1 and 100.

```python
prime_nums=[]

def primes():
    nums = 100
    def is_prime(num):
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                return False
        return True
    for num in range(1, nums):
        if is_prime(num):
            prime_nums.append(num)
    return prime_nums

def main():
    print(primes())
main()
```
