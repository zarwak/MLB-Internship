Student Performance Analysis


```python
import pandas as pd
```

Load the dataset using Pandas.

```python
df = pd.read_csv("student_performance.csv")
pd.concat([df.head(), df.tail()])
```

```python
print("\nSummary Statistics (subjects):")
print(df[['Python','Mathematics','Statistics','Machine_Learning']].describe())
```

Display basic information about the dataset.

```python
df.info()
```

```python
df.describe()
```

```python
print(df.dtypes)
```


Calculate average marks for each subject.

```python


subject_cols = ['Python','Mathematics','Statistics','Machine_Learning']
avg_marks = df[subject_cols].mean()
print("Average marks per subject:")
print(avg_marks)

```

Identify the top 5 performing students.
Find students scoring below the average.
Display the total number of students.
Save the cleaned or analyzed dataset as a new CSV file.

```python
df['Avg_Score'] = df[subject_cols].mean(axis=1)
top5 = df.nlargest(5, 'Avg_Score')[['Student_ID','Name','Avg_Score']]
print("Top 5 performing students:")
print(top5)
```

Find Students Scoring Below the Average

```python
overall_avg = df['Avg_Score'].mean()
below_avg = df[df['Avg_Score'] < overall_avg]
print(f"Overall average score: {overall_avg:.2f}")
print(f"Students below overall average ({len(below_avg)} students):")
print(below_avg[['Student_ID','Name','Avg_Score']])
```

Display Total Number of Students

```python
print(f"Total number of students: {len(df)}")
```

Save the Processed Dataset

```python
df['Above_Avg'] = df['Avg_Score'] >= overall_avg
processed_df = df.copy()
processed_df.to_csv('student_analysis_processed.csv', index=False)
print("Processed dataset saved as 'student_analysis_processed.csv'")
```
