## 1. Series and DataFrames

Pandas has exactly two objects you need to know:

* **Series** = one column of data + an **index** (the labels down the side). Think of it as a
  NumPy array that remembers what each row is called.
* **DataFrame** = a whole table = several Series glued together, sharing one index.

The index is the big idea: pandas keeps labels attached to your data, so rows never get
silently mixed up.

```python
# A Series from a list - pandas creates a default index 0, 1, 2, ...
s = pd.Series([10, 20, 30, 40])
print(s)
print("\nvalues:", s.values)   # the underlying NumPy array
print("index :", list(s.index))
```

```python
# A Series with your own labels - now you can look values up by name.
marks = pd.Series([88, 92, 79], index=['Ali', 'Sara', 'Zara'], name='marks')
print(marks)
print("\nSara's mark:", marks['Sara'])

# Maths works element-wise, exactly like NumPy.
print("\nafter +5 bonus:\n", marks + 5)
```

```python
# A DataFrame from a dictionary: each key becomes a column name, each value a column.
df = pd.DataFrame({
    'name':   ['Ali', 'Sara', 'Zara', 'Omar'],
    'age':    [21, 22, 20, 23],
    'city':   ['Lahore', 'Karachi', 'Lahore', 'Islamabad'],
    'score':  [88, 92, 79, 85]
})
df      # last line of a cell displays nicely as a table
```

```python
# One column of a DataFrame IS a Series - this is why Series matter.
print(type(df['score']))
print(df['score'])

# Two or more columns (note the double brackets) give you a smaller DataFrame.
print(type(df[['name', 'score']]))
```

## 2. Reading CSV files

Real data usually arrives as a **CSV** (comma-separated values) - a plain text file where
each line is a row and commas separate the columns.

`pd.read_csv()` turns that file into a DataFrame in one line. First we create a sample file
so this notebook runs anywhere.

```python
# Create a small sample dataset and save it as a CSV file (this is our stand-in for real data).
np.random.seed(0)

students = pd.DataFrame({
    'student_id': range(1, 16),
    'name': ['Ali','Sara','Zara','Omar','Hina','Bilal','Ayesha','Usman','Fatima','Hamza',
             'Noor','Saad','Iqra','Danish','Mariam'],
    'age': [21,22,20,23,21,24,22,20,23,21,22,25,20,23,21],
    'city': ['Lahore','Karachi','Lahore','Islamabad','Karachi','Lahore','Islamabad','Karachi',
             'Lahore','Islamabad','Karachi','Lahore','Islamabad','Karachi','Lahore'],
    'course': ['ML','ML','Data','Web','ML','Data','Web','ML','Data','Web','ML','Data','ML','Web','Data'],
    'score': [88,92,79,85,95,61,73,90,58,84,77,69,99,45,81],
    'attendance': [95,88,76,91,99,60,72,94,55,89,80,70,98,40,85]
})

students.to_csv('students.csv', index=False)   # index=False -> do not save the 0,1,2 column
print("students.csv written")
```

```python
# The one line that matters: read the file back into a DataFrame.
df = pd.read_csv('students.csv')
df.head()      # head() shows the first 5 rows - always check the file loaded correctly
```

```python
# Useful read_csv options you will meet sooner or later (shown, not all needed here):
#   sep=';'                 -> a different separator
#   index_col='student_id'  -> use an existing column as the index
#   usecols=['name','score']-> load only some columns (saves memory on big files)
#   nrows=100               -> peek at the first 100 rows of a huge file
#   na_values=['?', 'NA']   -> treat these strings as missing values

peek = pd.read_csv('students.csv', usecols=['name', 'score'], nrows=3)
peek
```

## 4. Selecting rows and columns

Three tools, and it is worth being clear about which is which:

* `df['col']` - grab a **column** by name.
* `df.loc[rows, cols]` - select by **label** (the row's index label / the column's name).
  The end of a `loc` slice **is included**.
* `df.iloc[rows, cols]` - select by **integer position**, like NumPy. The end **is excluded**.

```python
# --- Columns ---
print(df['name'].head(3))                # one column -> Series
df[['name', 'score', 'city']].head(3)    # several columns -> DataFrame (double brackets)
```

```python
# --- Rows by position: iloc ---
print(df.iloc[0])        # the very first row, shown as a Series
print()
df.iloc[0:3]             # rows 0,1,2 - the end is EXCLUDED, just like Python slicing
```

```python
# iloc[rows, columns] - both by position
print(df.iloc[0, 1])         # row 0, column 1
print(df.iloc[0:3, 1:4])     # first 3 rows, columns 1-3
print(df.iloc[-1])           # last row
```

```python
# --- Rows by label: loc ---
# Our index is still 0,1,2..., so labels look like positions here - but note 0:3 gives 4 rows.
print(df.loc[0:3, ['name', 'score']])

# Set a meaningful index and the difference becomes obvious.
df_named = df.set_index('name')
print(df_named.loc['Sara'])                       # look a student up by name
print(df_named.loc[['Sara', 'Ali'], ['score', 'city']])
```

```python
# Handy shortcuts
print(df['score'].head())      # column as a Series
print(df.at[2, 'name'])        # a single cell, by label - fastest option
print(df.iat[2, 1])            # a single cell, by position
```

## 5. Filtering data

Filtering = keeping only the rows where a condition is `True`.

Write the condition first (it produces a column of True/False), then put it inside `df[...]`.
Combine conditions with `&` (and) and `|` (or) - each condition needs **its own brackets**,
because `&` binds more tightly than `>` in Python.

```python
# Step 1: the condition on its own is just a True/False Series.
print((df['score'] > 85).head())

# Step 2: use it to keep matching rows.
df[df['score'] > 85]
```

```python
# Combining conditions - brackets around each one are required.
high_and_present = df[(df['score'] > 80) & (df['attendance'] > 85)]
print("scored >80 AND attended >85%:")
print(high_and_present[['name', 'score', 'attendance']])

# | means OR
weak = df[(df['score'] < 60) | (df['attendance'] < 60)]
print("\nat-risk students:")
print(weak[['name', 'score', 'attendance']])
```

```python
# isin() - match against a list of values (much cleaner than chaining several ORs).
df[df['city'].isin(['Lahore', 'Karachi'])].head()
```

```python
# between() for ranges (both ends included), and ~ to invert any condition.
print(df[df['age'].between(21, 22)][['name', 'age']])
print("\nNOT from Lahore:")
print(df[~(df['city'] == 'Lahore')][['name', 'city']].head())
```

```python
# Filtering text with .str - use .str.contains / .startswith / .lower first.
print(df[df['name'].str.startswith('S')][['name', 'course']])

# query() is an alternative style some people find more readable.
print(df.query('score > 90 and city == "Karachi"'))
```

```python
# Filter, then sort, then take the top few - a very common combination.
top3 = df[df['course'] == 'ML'].sort_values('score', ascending=False).head(3)
top3[['name', 'course', 'score']]
```

## 6. Basic statistics

`describe()`, `info()` and `head()` are the three you will type most often:

* `head()` - *what does the data look like?*
* `info()` - *what types are there, and is anything missing?*
* `describe()` - *what are the numbers actually doing?*

```python
# describe() summarises every numeric column at once:
#   count = non-missing values     mean/std = average and spread
#   min/max = range                25%/50%/75% = quartiles (50% is the median)
df.describe()
```

```python
# include='object' summarises the TEXT columns instead:
#   unique = number of distinct values, top = most common, freq = how often it appears
df.describe(include='object')
```

```python
# Individual statistics on one column, when you need just one number.
print("mean score  :", df['score'].mean())
print("median score:", df['score'].median())
print("std dev     :", df['score'].std())
print("min / max   :", df['score'].min(), "/", df['score'].max())
print("total       :", df['score'].sum())
print("count       :", df['score'].count())
```

```python
# Run a statistic across every numeric column at once.
print(df[['age', 'score', 'attendance']].mean())

# Correlation: how strongly do two columns move together? (+1 together, -1 opposite, 0 unrelated)
print("\ncorrelation:")
print(df[['age', 'score', 'attendance']].corr())
```

```python
# groupby = split the data into groups, apply a statistic, combine the results.
print("average score per course:")
print(df.groupby('course')['score'].mean())

print("\nseveral statistics at once:")
print(df.groupby('city')['score'].agg(['count', 'mean', 'max']))
```

```python
# Putting it all together: the routine to run on any new dataset.
print("1. SHAPE   ->", df.shape)
print("\n2. HEAD"); print(df.head(3))
print("\n3. INFO");  df.info()
print("\n4. MISSING"); print(df.isnull().sum())
print("\n5. DESCRIBE"); print(df.describe())
```

## Quick recap

| Task | Code |
|---|---|
| Build data | `pd.Series([...])`, `pd.DataFrame({...})` |
| Load a file | `pd.read_csv('file.csv')` |
| First look | `.head()`, `.tail()`, `.shape`, `.columns` |
| Health check | `.info()`, `.isnull().sum()`, `.dtypes` |
| Pick columns | `df['col']`, `df[['a','b']]` |
| Pick rows | `df.loc[label]` (by name), `df.iloc[0:3]` (by position) |
| Filter | `df[df['score'] > 80]`, `&`, `|`, `.isin()`, `.between()` |
| Summarise | `.describe()`, `.mean()`, `.value_counts()`, `.groupby()` |

Together with the NumPy notebook, this is the toolkit every ML project starts with.
