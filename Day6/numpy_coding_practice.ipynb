## NumPy



## Create 1D and 2D arrays.

```python
import numpy as np
```

```python
list1 = [1, 2, 3, 4, 5]
list2 = [[6, 7], [8, 9]]

arr1 = np.array(list1)
arr2 = np.array(list2)

print(arr1)
print(arr2)
```

## Perform arithmetic operations on arrays.

```python
arr = arr1 + 10
print(arr)

arr = arr2 * 2
print(arr)

arr = arr1 - 2
print(arr)

arr = arr2/2
print(arr)
```

## Find the maximum, minimum, mean, and sum of an array.

```python
arr = np.arange(1,100,3)
mx = arr.max()
mn = arr.min()
arr_mean = arr.mean()
arr_sum = arr.sum()

print(f'maximum : {mx} \nminimum: {mn} \nmean: {arr_mean}\nsum: {arr_sum} \n')
```

## Reshape arrays into different dimensions.

```python
arr = np.linspace(0,5,10)
arr1 = arr.reshape(2,5)

arr2 = np.random.randint(5, size=(2,3))
arr3 = arr2.reshape(3,2)

print(arr)
print(arr1)
print(arr2)
print(arr3)
```

## Slice and index arrays.

```python
arr1 = np.linspace(0,5,10)
print(arr1)
print(arr1[0])
print(arr1[-1])
print(arr1[::-1])
print(arr1[arr1>3])
print(arr1[::3])

arr2 = np.random.randint(5, size=(2,3))
print(arr2)
print(arr2[0])
print(arr2[-1])
print(arr2[0][0])
print(arr2[-1][-1])
print(arr2[::-1])
print(arr2[arr2>3])
print(arr2[::3])
```
