## Create a text file and write data into it.

```python
with open('textfile.txt','w') as textfile:
    textfile.write("Hi there! Its my internship task.")
```

```python
## Read and display file contents.
```

```python
with open('textfile.txt','r') as textfile:
    contents = textfile.read()
    print(contents)
```

## Append new data to an existing file.

```python
with open('textfile.txt','a') as textfile:
    textfile.write("\nI am learning file handling in python.")
```

## Count the number of lines in a file.

```python
with open('textfile.txt','r') as textfile:
    contents = textfile.readlines()
    print(len(contents))
```
