## Store student information in a JSON file.

```python

import json

information = {
    "name": "zarwa",
    "roll_no": 17,
    "course": "DS"
}

json.dump(information, open('out.json', 'w'))
```

## Read data from a JSON file.

```python
json_data = json.load(open('out.json', 'r'))
print(json_data)
```

## Update an existing student's information.

```python
import json


try:
    with open('out.json', 'r') as f:
        students = json.load(f)
except (FileNotFoundError, json.JSONDecodeError):
    students = []

roll_to_update = int(input("Enter roll number of student to update: "))
new_course = input("Enter new course: ")

updated = False
for student in students:
    if student.get('roll_no') == roll_to_update:
        student['course'] = new_course   
        updated = True
        break

if updated:
    
    with open('out.json', 'w') as f:
        json.dump(students, f, indent=2)
    print("Student updated successfully!")
else:
    print(f"No student found with roll number {roll_to_update}.")
```

## Add a new student to the JSON file.

```python
import json

try:
    with open('out.json', 'r') as f:
        data = json.load(f)
except (FileNotFoundError, json.JSONDecodeError):
    print(" File missing ")
    data = []   # This discards the corrupted data

# Now proceed to add a new student
name = input("Enter name: ")
roll_no = int(input("Enter roll number: "))
course = input("Enter course: ")
new_student = {
    "name": name,
    "roll_no": roll_no,
    "course": course
}
data.append(new_student)

with open('out.json', 'w') as f:
    json.dump(data, f, indent=2) 
```
