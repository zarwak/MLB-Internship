Person Class

```python
class Person:
    def __init__ (self, name,age,address):
        self.name = name
        self.age = age
        self.address = address

    def introduction(self):
        print(f"Name:{self.name}\nAge:{self.age}\nAddress:{self.address}")
```

Student class

```python
class Student(Person):
    def __init__(self,  name, age, address,student_id, grade, courses= None):
        super().__init__(name=name, age=age, address=address)
        self.student_id = student_id
        self.grade = grade
        self.courses = courses if courses is not None else []

    def display(self):
        super().introduction()
        print(f"I am a student with ID: {self.student_id} and  Grade: {self.grade}")

    def upgrade_grade(self, new_grade):
        self.grade = new_grade
        print(f"{self.name}'s grade has been upgraded to {self.grade}.")

    def add_course(self, course):
        if course in self.courses:
            print(f"{course} is already enrolled for {self.name}.")
        self.courses.append(course)
        print(f"{course} has been added to {self.name}'s courses.")
        
```

Teacher class

```python
class Teacher(Person):
    def __init__(self, name,age, adress, employee_id, subject, salary):
        super().__init__(name, age=None, address=None)
        self.employee_id = employee_id
        self.subject = subject
        self.salary = salary

    def display(self):
        super().introduction()
        print(f"I am a teacher with ID: {self.employee_id}, Subject: {self.subject}, and Salary: {self.salary}")

    def raise_salary(self, amount):
        self.salary += amount
        print(f"{self.name}'s salary has been increased by {amount}. New salary: {self.salary}.")
```

Main class

```python


def main():
    print("INHERITANCE DEMO\n")

    # Create a Person (base class)
    p = Person("John Doe", 45, "123 Main St")
    p.introduction()
    print()

    # Create a Student
    s = Student("Alice", 20, "456 Oak Ave", "S123", "A", ["Math", "Physics"])
    s.introduction()
    s.add_course("Chemistry")
    print()

    # Create a Teacher
    t = Teacher("Dr. Smith", 50, "789 Pine Rd", "T456", "Computer Science", 80000)
    t.introduction()
    t.raise_salary(5000)



main()
```
