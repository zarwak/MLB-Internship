student class

```python
class Student:
    def __init__(self, name, id, grade):
        self.name = name
        self.id=id
        self.grade=grade
    
    def display(self):
        print(f"Student: {self.name}, Student_ID: {self.id}, Grade: {self.grade}")

    def upgrade_grade(self,new_grade):
        self.grade = new_grade
        print(f"Grade updated to {self.grade}")
        
```

Employee class

```python
class Employee:
    def __init__(self,name,emp_id,dept,sal):
        self.name=name
        self.emp_id=emp_id
        self.dept=dept 
        self.sal=sal

    def display(self):
        print(f"Name:{self.name}, Employee_id: {self.emp_id}, Department:{self.dept},Salary:{self.sal}")
```

Car class

```python
class Car:
    def __init__(self,make,model,year,colour,distance_covered=0):
        self.make = make
        self.model = model
        self.year = year
        self.colour = colour
        self.distance_covered = distance_covered
        self.engine_running = False

    def start_engine(self):
        if not self.engine_running:
            self.engine_running = True
            print(f"{self.make} {self.model} engine started. ")
        else:
            print("Engine is already running. ")

    def stop_engine(self):
        if self.engine_running:
            self.engine_running = False
            print(f"{self.make} {self.model} engine stopped. ")
        else:
            print("Engine is already OFF. ")

    def drive(self, distance):
        if self.engine_running:
            self.distance_covered += distance
            print(f"Drove {distance} miles. Total mileage: {self.distance_covered}")
        else:
            print("Start the engine first!")

    def paint(self, new_color):
        self.colour = new_color
        print(f"Car painted {new_color}.")

    def display(self):
        print(f"{self.year} {self.make} {self.model} ({self.colour}), Mileage: {self.distance_covered}")
```

Multiple objects from same class

```python

def main():
    print("="*40)
    print("STUDENT DEMO")
    s1 = Student("Alice", "S001", "A")
    s2 = Student("Bob", "S002", "B")
    s1.display()
    s2.display()
    s1.upgrade_grade("A+")
    s1.display()

    print("\n" + "="*40)
    print("EMPLOYEE DEMO")
    e1 = Employee("Charlie", "E101", "Engineering", 75000)
    e2 = Employee("Diana", "E102", "Marketing", 68000)
    e1.display()
    e2.display()
    e1.display()

    print("\n" + "="*40)
    print("CAR DEMO")
    car1 = Car("Toyota", "Camry", 2020, "Red", 15000)
    car2 = Car("Honda", "Civic", 2022, "Blue")
    car1.display()
    car2.display()
    car1.start_engine()
    car1.drive(100)
    car1.paint("Black")
    car1.display()
    car1.stop_engine()

main()
```
